```python
df_2022 = data[['Country Name', '2022']].dropna()
df_2022.columns = ['Country', 'Population_2022']
```

```python
# Plotting the bar chart for population distribution by country in 2022
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='Population_2022', data=df_2022.sort_values(by='Population_2022', ascending=False).head(10))
plt.xticks(rotation=90)
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
# Plotting the histogram for population distribution in 2022
plt.figure(figsize=(14, 8))
sns.histplot(df_2022['Population_2022'], bins=30, kde=True)
plt.title('Distribution of Population in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
